```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
gran = pd.read_csv("./destinations.csv", encoding="latin1")
clima = pd.read_csv("./top100cities_weather_data.csv", encoding="latin1")
features = pd.read_csv("./travel_destinations.csv", encoding="latin1")

unio = pd.merge(gran,clima, on='Destinations',how='inner')
unio2 = pd.merge(unio,features, on='Destinations',how='inner')

unio2.columns
```

```python
db = unio2.drop(columns=['Latitude_y','Longitude_y','Country','Country_y','Region', 'Best Time to Visit', 'Category_x', 'Majority Religion', 'Category_y'])

db['Category_y']
```

```python
vectorizer = TfidfVectorizer(max_features=20) # Agafa les 20 + importants
X = vectorizer.fit_transform(db['Category_y'])

category_features = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

db_nova = pd.concat([db.reset_index(drop=True), category_features], axis=1)

db_nova
```

```python
categories_grup = {
    "Arquitectura_Monuments": ["architecture", "architecture (Gaudi)", "architecture (Grand Place)", "architecture (Moorish)", "castle", "palaces", "landmarks", "unique cityscape"],
    "Art_Cultura": ["art", "art (Renaissance)", "museums", "theatre", "music", "flamenco", "Fado music", "comics", "culture", "design"],
    "Gastronomia_Begudes": ["food", "food (Nicoise)", "food (tapas)", "chocolate", "coffee houses", "beer", "beer culture (Oktoberfest)", "wine"],
    "Historia_Tradicio": ["ancient ruins", "history", "mythology", "religion", "royalty"],
    "Naturalesa": ["French Riviera", "lake", "islands (nearby)", "parks", "hiking", "cycling", "coastline", "archipelago"],
    "Platja": ["beach"],
    "Events_Festvals": ["festivals (Aug)", "festivals (Semana Santa, Feria)"],
    "Compres": ["shopping", "bazaars", "leather goods", "hygge"],
    "Nightlife_Entreteniment": ["nightlife", "promenade"],
    "Ciutat_EstructuraUrbana": ["Old Town", "canals", "gondolas", "viewpoints"],
    "Especific": ["East meets West", "EU capital", "finance", "romance"],}


db["Category_y"] = db["Category_y"].str.split(", ")

for group_name, group_categories in categories_grup.items():
    db[group_name] = db["Category_y"].apply(
        lambda cats: 1 if any(cat in group_categories for cat in cats) else 0
    )

db = db.drop(columns=["Category_y"])

db.to_csv("db.csv", index=False, encoding="utf-8")
```
